# Bivariate Visualizations
- Visualizations of two variables.
- We use bivariate visualizations to look at relationships.

- The main plot types of bivariate visualizations:
    1. `Scatter plot, heatmap, line plot`
        - For quantitative variable vs. quantitative variable.
    2. `Violin plot, box plot`
        - For quantitative variable vs. qualitative variable.
    
    3. `Clustered bar chart, categorical heatmap`
        - For qualitative variable vs. qualitative varialbe.

## Scatter Plot

- Visualize the relationship between two **quantitative** variables.
- The values of one variable are indicated by position on the X axis, and the values of the other variable are indicated by position of the Y axis.
- One point is plotted for every observation we have in our data, resulting in a cloud of points.
- The pattern of the points can show us the direction and the strength of the relationship.

### Pearson Correlation Coefficient (r)
- A statistic quantifying the strength of **linear** correlation between two numeric variables.
- The Pearson correlation coefficient only captures **liner** relationships.
- **-1<=r<=1**.
- If the value of r is positive, this indicates a positive relationship, and a negative r value indicates a negative relationship.
- If r is closer to 1 or -1, this indicates a stronger and more predictable relationship.
- The correlation coefficient doesn't necessarily tell you anything about the steepness of a line that models the relationship.
- It's important to consider not just the numbers alone, but whether or not they are meaningful to your data.
- In **Anscombe's quartet**: Multiple datasets with the same r value, but when plotted, demonstrated very different shape characteristics.
- If one of our goals is to create a regression model around the data, scale transformations can be very useful to change a non linear trend in the natural units into a linear trend in the transformed units. If the transformed units form a linear pattern, try using the value of (X) to predict the log of (Y) in a regression approach.

### Overplotting, Transparency and Jitter

- Overplotting occurs in cases where there are an extremely large number of points, or the numeric data is **discrete** in nature, which results in a an undetailed plot where there are too many overlapping points in a small area, making it difficult to interpret the plot.
- Overplotting can be resolved with various techniques:
    - Sampling
    - Transparency
    - Jitter

1. Sampling
    - Random sampling to plot a small number of points.
    - The random selection of points gives us a good intuition for the general distribution of data.
2. Transparency
    - Plot all the points with transparency.
    - Each individual point will be lighter in color, but when points overlap, the overlapped region becomes darker.
    - The darkest regions, with the highest number of points, indicate the peaks of the distribution.
    - Since a higher density results in a darker color, we have, in an implied way, used color to augment what would've been obsecured from using position alone.
3. Jitter
    - Used when one or both of the variables are discrete.
    - Jittering adds a small amount of random noise to the position of each data point, so that multiple data points that have the same values are now spread over a small area to prevent overplotting.
    - As long as the jitter keeps each of the discrete values in distinct bands, this can be a big help in the identification and interpretation of trends.

```python
#sampling
locs = np.random.choice(range(fuel_econ['comb'].count()), 300, replace = False)
sampled_data = fuel_econ.loc[locs, ['displ', 'comb']]
plt.scatter(data = sampled_data, x= 'displ', y = 'comb');
```

```python
sb.regplot(data = fuel_econ, x = 'year', y = 'comb')
plt.xticks([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]);
```

The regression line suggests a slight upward trend in efficiency, But this is not clear from the data.

```python
#x values will be plotted in a uniform +- 0.3 range of their true values
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3)
plt.xticks([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]);
```

- The **x_jitter** parameter only works in **sb.regplot** function, and not plt.scatter.

```python
#Alter the transparency of the plotted points
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3, scatter_kws = {'alpha' : 1/20})
plt.xticks([2012, 2014, 2016, 2018, 2020]);
```

- `scatter_kws` parameter: Additional keyword arguments to pass to `plt.scatter` and `plt.plot`.
- When using **plt.scatter** function instead, the **alpha** parameter can be added directly to the function call.
- Through all of this, the regression line doesn't change at all. The jitter and transparency don't affect the underlying values for the purposes of regression line fits.

### Heat Maps
- Numeric heatmaps
- Categorical heatmaps

- A heat map is **a 2-d version of the histogram** and it can be used as an alternative to a scatter plot.
- It is a bivariate plot type that exploits the relationship between color and point density to visualize the major trends in the data.
- It is favored over scatter plots when you have two **discrete** variables.
    - jittered scatter plots can be imprecise.
    - A heat map is an alternative to adding transparency to a scatter plot when you have a very large amount of data.
- In a heat map, the field is divided into a grid of cells, and the number of data points in each cell is counted up. Then each cell is given a color based on the count of values.
- You need to be aware of the impact of bin size on the interpretation of the data. Too large a bin size, and you may fail to see the major trends in the data, and too small a bin size, you may end up distracted by the noise.
- Heatmaps can also be used as **2-d versions of bar charts**, plotting counts on combinations of two categorical variables instead of numeric variables. There is a function in seaborn, **heatmap**, that is built for categorical heatmaps.

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb')
plt.colorbar()
#to add a color bar to the plot
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

Similar to the univariate **hist** function, each dimension is split into 10 bins.

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5, cmap = 'viridis_r')
plt.colorbar()
#to add a color bar to the plot
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

- **cmin**: specifies the minimum value in a cell before it will be plotted and colored.
- setting cmin = 0.5, any area with no counts will not be plotted.
- **camp**: used to set the color map used in the plot.
- adding '_r' to the name of the color map indicates a reverse color map so that darker color is associated with a higher count.

```python
fuel_econ[['displ', 'comb']].describe()
```

```python
bins_x = np.arange(0.6, 7 + 0.3, 0.3)
bins_y = np.arange(12, 58 + 3, 3)
h = plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5, cmap = 'viridis_r', bins = [bins_x, bins_y])
plt.colorbar()
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

- The bins parameter takes as an argument a list of two lists specifying the bin edges for the x and y axes in that order.

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb', fit_reg = False, scatter_kws = {'alpha' : 1/20}, 
          x_jitter = 0.1, y_jitter = 2)
plt.gca().set_title('Scatter Plot')
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)')
plt.subplot(1,2,2)
bins_x = np.arange(0.6, 7 + 0.3, 0.3)
bins_y = np.arange(12, 58 + 3, 3)
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5, cmap = 'viridis_r', bins = [bins_x, bins_y])
plt.colorbar()
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

- Since assigning values based on color is imprecise, you should consider adding **count annotations** on top of each grid cell, especially if you only have a limited number of cells.

```python
h2d = plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5, cmap = 'viridis_r')
plt.colorbar()
counts = h2d[0]
bins_x = h2d[1]
bins_y = h2d[2]
for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i][j]
        if c >= 7:
            plt.text(bins_x[i] + 0.3, bins_y[j] + 1.5, int(c), color = 'w', ha = 'center', va = 'center')
        elif c > 0:
            plt.text(bins_x[i] + 0.3, bins_y[j] + 1.5, int(c), color = 'b', ha = 'center', va = 'center')
```

If you have too many cells in your heatmap, then the annotations will end up be too overwhelming, too much to attend to. In cases like this, it's best to leave off the annotations and let the data and colorbar speak for themselves. You are more likely to use annotations in categorical heatmaps, where there are going to be fewer cells plotted.

## Violin Plot

- Plot a qualitative variable against a quantitative variable.
- In violin plot, instead of plotting each data point individually, we have a curve for each level of the categorical variable whose area depicts the distribution of data points (the distribution is plotted as a kernel density estimate).
- For each level of the categorical variable, a distribution of the values of the numeric variable is plotted.
- Where the curve is wider, there are more data points.
- For each level of the categorical variable, an area is plotted to show the approximate density of data points.

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = sedan_classes)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)
```

In the above code, `VClass` column is converted from a plain object type into an ordered categorical type, putting the classes in order of size.

```python
basecolor = sb.color_palette()[0]
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', color = basecolor)
plt.xticks(rotation = 15);
```

- Inside each violin, there is a miniature box plot.

```python
#inner = None so that the violin don't display the inner box plots
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', color = basecolor, inner = None)
plt.xticks(rotation = 15);
```

- The largest cars seem to have the lowerst fuel efficiency, but the long tails suggest outliers with the highest of efficiencies.
- Compact and Midsize cars have the best efficiencies.
- The violin plot can be rendered horizonatally by setting the Y axis to receive the categorical variable.

```python
sb.violinplot(data = fuel_econ, x = 'comb', y = 'VClass', color = basecolor, inner = None);
```

- Seaborn is smart enough to make an inference on which orientation is requested, depending on whether X or Y recieves the categorical variable. But if both variables are numeric (one is discretely valued), then the **orient** parameter can be used to specify the orientation.

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3, fit_reg = False, scatter_kws = {'alpha' : 1/10});
plt.subplot(1,2,2)
sb.violinplot(data = fuel_econ, x = 'year', y = 'comb', color = basecolor);
```

To plot a qualitative variable against a quantitative variable using a scatter plot, map each level of the categorical variable to an arbitrary number, then use a jittered scatter plot to visualize the qualitative variable, represented with numbers, against a quantitative variable.

```python
plt.figure(figsize = (15, 5))
plt.subplot(1,2,1)
class_num = {'Minicompact Cars' : 1, 'Subcompact Cars' : 2, 'Compact Cars' : 3, 'Midsize Cars' : 4, 'Large Cars' : 5}
fuel_econ['ClassNumber'] = fuel_econ['VClass'].map(class_num)
fuel_econ['ClassNumber'] = fuel_econ['ClassNumber'].astype(np.int)
sb.regplot(data = fuel_econ, x = 'ClassNumber', y = 'comb', fit_reg = False, x_jitter = 0.3, scatter_kws = {'alpha' : 1/10})

plt.subplot(1,2,2)
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', color = basecolor)
plt.xticks(rotation = 15);
```

## Box Plot

- In a box plot, descriptive statistics are computed within each level of the categorical variable, and then depicted graphically using box and whiskers.
- 25% of the data falls between each set of horizontal lines.
- Box plots often have outliers plotted as points beyond the end of whiskers.
- The most common upper bound on the length of the whiskers is 1.5 times the interquartile range (IQR) or box length.

- Example dataset:
    - Q1 = 53
    - Q2 = 55
    - Q3 = 58.5
    - IQR = 5.5
    - MIN = 49
    - MAX = 68
- Maximum whisker length = 1.5 * IQR = 1.5 * 5.5 = 8.25
- If we have outliers that are further than this distance above the top of the box or below the bottom of the box, they are depicted as single points and the whiskers are only extended to the farthest point within the 1.5 times IQR.
- The maximum bound for the upper whisker = Q3 + 8.25 = 58.5 + 8.25 = 66.75
- So the maximum value of 68 is depicted as an outlier, and the whisker is placed at the next large value of 63.

- Although there are some distributional details that can get lost with a box plot, it is a cleaner summary of the data, and it's easier to compare statistics between the groups with a box plot.
- So when it comes to Explanatory data visualizations or if you have a lot of groups to compare, you might want to choose a box plot for simplicity and focus.
- Because the box plot lacks as nuanced a depiction of distributions as the violin plot, the violin plot maybe a better option for exploration of data, especially since Seaborn's implementation also includes the box plot by default.
- Violin plot: better for exploratory visuals, Box plot: better for explanatory visuals.

```python
sb.boxplot(data = fuel_econ, x = 'VClass', y = 'comb', color = basecolor)
plt.xticks(rotation = 15);
```

- The bottom of each box indicates the first quartile (Q1), and the top of each box indicates the third quartile (Q3). The central line is the median of each group.
- Whiskers brach out to the min and max of each group with outliers depicted as individual points past a maximum whisker length that depends on each group's box height. The taller each box, the wider the whiskers can be.
- To remove the outliers from the chart, set **showfliers = False**.

```python
sb.boxplot(data = fuel_econ, x = 'VClass', y = 'comb', color = basecolor, showfliers = False)
plt.xticks(rotation = 15);
```

```python
plt.figure(figsize = (20, 5))
plt.subplot(1,3,1)
ax1 = sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', color = basecolor)
plt.xticks(rotation = 15)

plt.subplot(1,3,2)
#plot the quartiles as dashed lines
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb', color = basecolor, inner = 'quartile')
plt.xticks(rotation = 15)
plt.ylim(ax1.get_ylim())

plt.subplot(1,3,3)
sb.boxplot(data = fuel_econ, x = 'VClass', y = 'comb', color = basecolor)
plt.xticks(rotation = 15)
plt.ylim(ax1.get_ylim());
```

- The **ylim** expression has been added to match the three plots' y-axis limits.
- The inner boxes and lines in the violin plot match up with the boxes and whiskers in the box plot.

```python
#The box plot can also be rendered horizontally
sb.boxplot(data = fuel_econ, x = 'comb', y = 'VClass', color = basecolor);
```

## Clustered Bar Chart

- A clustered bar chart is used to look at the relationship between two categorical variables.
- Heatmaps can also be used as 2-d versions of bar charts, plotting counts on combinations of two categorical variables. But if we want to avoid the imprecision of using color to encode numbers, then the standard plot of choice is the clustered bar chart.
- A clustered bar chart starts out the same as the univariate bar chart with the levels of one categorical variable on the X-axis and counts to be plotted on the Y axis. However, instead of just one bar being plotted at each X position, we instead have multiple bars, one for each level of the second categorical variable.
- Typically we apply a color or texture encoding to each bar in each cluster to distinguish between the levels of the second categorical variable.
- A clustered bar chart is more easily interpreted than the heatmap.

```python
#The two main types of transmission: Automatic, Manual
fuel_econ['trans_type'] = fuel_econ['trans'].map(lambda x : x.split()[0])
```

The sb.heatmap() function takes as an argument a 2d array of df with the values to be depicted. So we need to summarize the data into a matrix that will be plotted.

```python
ct_counts = fuel_econ.groupby(['VClass', 'trans_type']).size().reset_index(name = 'count')
ct_counts = ct_counts.pivot(index = 'VClass', columns = 'trans_type', values = 'count')
ct_counts
```

```python
sb.heatmap(ct_counts);
```

```python
#Add annotations to each cell as integers
sb.heatmap(ct_counts, annot = True, fmt = 'd');
```

```python
sb.countplot(data =fuel_econ, x = 'VClass', hue = 'trans_type')
plt.xticks(rotation = 15);
```

```python
fuel_econ.groupby('trans_type').size()
```

```python
#Further customization of the plot legend ....> Documentation
sb.countplot(data = fuel_econ, x = 'VClass', hue = 'trans_type')
plt.xticks(rotation = 15)
plt.legend(loc = 0, framealpha = 1, title = 'Transmission');
```

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
sb.heatmap(ct_counts, annot = True, fmt = 'd')
plt.subplot(1,2,2)
sb.countplot(data = fuel_econ, x = 'VClass', hue = 'trans_type')
plt.xticks(rotation = 15)
plt.legend(loc = 0, framealpha = 1, title = 'Transmission');
```

## Faceting


- Creating multiple copies of the same type of plot but for different subsets of the data. Faceting is used to subset your dataset across levels of a categorical varible and then create one plot for each subset.
- One way that you could think about a clustered bar chart is that it's a set of side-by-side univariate bar charts on different subsets of the data.
- Using faceting, univariate plots can be adopted to view bivariate relationships.
- Faceting can be used for any type of plot.
- E.g. a violin plot can be thought of as very similar to a set of histograms faceted by levels of a categorical variable.
- Faceting is good for breaking down potentially complex patterns into simple parts. This is especially useful if you have a categorical variable with a lot of levels. For example, user registeration dates can be faceted by month and arranged in a 3x4 grid.
- When faceting, **axis scales and limits** should be consistent across each subplot.

```python
sb.violinplot(data =fuel_econ, x = 'VClass', y = 'comb', color = basecolor)
plt.xticks(rotation = 15);
```

A violin plot can be thought of as very similar to a set of side by side histograms faceted by the levels of a categorical variable.

```python
g= sb.FacetGrid(data = fuel_econ, col = 'VClass')
#For whatever plot type we facet with, there will be a different plot for each level of the "VClass" column
g.map(plt.hist, 'comb');
```

- All of the data is plotted using the default 10 bins for the hist function, but the axis limits are the same for each facet.

```python
#Adding the bins parameter to make sure that the bins across each facet is the same
bin_edges = np.arange(12, 58+2, 2)
g = sb.FacetGrid(data = fuel_econ, col = 'VClass')
g.map(plt.hist, 'comb', bins = bin_edges);
```

```python
g = sb.FacetGrid(data =fuel_econ, col = 'VClass', col_wrap = 3)
g.map(plt.hist, 'comb', bins = bin_edges);
```

- **col_wrap** argument limits the number of facets that can be placed side by side. This is useful if you have a lot of levels in your faceted variable.
- **sharey** argument (default : True): All Y axis limits are the same across the different facets. If False, each facet can have its own Y axis limit.

```python
g = sb.FacetGrid(data = fuel_econ, col = 'VClass', col_wrap = 3, sharey = False)
g.map(plt.hist, 'comb', bins = bin_edges);
```

- Setting **sharey = False**, each facet can have its own Y axis limit.
- A failure to have the same limits on each subplot will make it easier to have misconceptions about the overall frequency or value range for each subplot of the data.
- It's a little easier this way to see the change in distribution across levels, but you also lose out on seeing the overall frequency of each group.
- If you're not paying attention, you might think that each class has a similar number of cars, when in fact, there are not exactly the same.

##### Other operations to increase the readability of the plots:
1. `height`: The height of each facet in inch.
2. `col_order`: Order the levels of the categorical variable.
3. `aspect`: Width of each facet = **aspect * height**.
4. Changing the title of each facet to just the categorical level name
    - `g.set_titles('{col_name})`

```python
g = sb.FacetGrid(data = fuel_econ, col = 'VClass', col_wrap = 3, height = 3, aspect = 1.5)
g.map(plt.hist, 'comb', bins = bin_edges)
g.set_titles('{col_name}');
```

## Adaptation of Univariate Plots
- Adopted Bar Chart (Qualitative Variable vs Quantitative Variable)
- Adopted Histogram (Quantitative Variable vs Quantitative Variable)

- In both histograms and bar charts, we had the frequency of data points on the vertical axis. If we instead reported the mean of a second variable, this would make the plot bivariate in nature.
- Sometimes we might add error bars to the plot to show the amount of uncertainty we have in the true average based on variance and sample size.
-  Even though the box plot takes the same amount of space as an adapted bar chart and reports more statistics, as an explanatory data visualization, the adapted bar chart is a more direct, and easier to read summary of your findings. For example, if the numeric variable in binary in nature, only taking values of 0 or 1, then a box plot or violin plot will not be informative, leaving the adapted bar chart as the best choice for displaying the data.
- Similarly, the adapted histogram acts as a condensed alternative to the scatter plot.

### Adapted Histogram

- A condensed alternative to the scatter plot.
- Matplotlib's **hist** function can also be adapted so that bar heights indicate value other than a count of points through the use of the **weights** parameter.
- By default, each data point is given a weight of 1, so that the sum of point weights in each bin is equal to the number of points.
- If we change the weights to be a representative function of each point's value on a second variable, then the sum will end up representing something other than a count.

```python
sb.regplot(data =fuel_econ, x = 'displ', y = 'comb', fit_reg = False);
```

```python
bin_edges = np.arange(0.6, 7.1 + 0.4, 0.4)
bin_indxs = pd.cut(fuel_econ['displ'], bins = bin_edges, right = False, include_lowest = True, labels = False).astype(int)
pts_per_bin = fuel_econ.groupby(bin_indxs).size()
comb_wts = fuel_econ['comb'] / pts_per_bin[bin_indxs].values
```

```python
plt.hist(data = fuel_econ, x = 'displ',bins = bin_edges, weights = comb_wts)
plt.xlabel('Displacement (L)')
plt.ylabel('Average Fuel Eff. (mpg)');
```

## Line plots

- In a line plot, we plot the relationship between two quantitative variables, one on each axis using a line.
- The Y value of the line at any given X value, reports the average value taken by the Y axis variable.
- In contrast to a scatter plot, where all data points are plotted, in a line plot, only one point is plotted for every unique X value or bin of X values (like a histogram). If there are multiple observations in an X bin, then the Y value of the point plotted in a line plot will be a summary statistic (like the mean or the median) of the data in the bin.
- A line plot is like a histogram with a variable average instead of a count on the Y axis, and point positions instead of bar heights.
- The use of lines over bars emphasizes two things:
    - We are more interested in **relative change**. So the need for a zero on the Y axis is not necessary.
    - The line emphasizes the sequential or connected nature of the X values. So it is inappropriate to use a line plot if we have a nominal variable on the X axis.
- If the X variable represents time, then a line plot of the data is frequently known as a **time series plot**.

- Line plots can be created using matplotlib's **errorbar** function, which plots a line between points with an optional setting for adding error bars.
- **sb.pointplot** : plots a qualitative variable against a quantitative variable.
- **plt.errorbar** : plots a quantitative variable against a quantitative variable.

```python
plt.errorbar(data = fuel_econ, x = 'displ', y = 'comb')
plt.xlabel("Displacement (L)")
plt.ylabel("Avg Combined Fuel Eff. (mpg)");
```

- The plot we get out is uninformative.
- All that matplotlib is doing is connecting each pair of x and y values in order as they appear in the dataframe without any sorting.
- A scatter plot is basically this plot without any connecting lines and distinct points at each value pair.

```python
plt.scatter(data = fuel_econ, x = 'displ', y = 'comb');
```

- To properly use errorbar function:
    - All of the data is sorted by the X variable.
    - There is only one Y value for each X value.

```python
bin_edges = np.arange(0.6, 7 + 0.2, 0.2)
```

```python
#bin center values make sure that the points are plotted in their accurate positions
bin_c = bin_edges[:-1]+0.1
```

```python
displ_bins = pd.cut(fuel_econ['displ'], bins = bin_edges, include_lowest = True)
```

```python
comb_mean = fuel_econ.groupby(displ_bins)['comb'].mean()
```

```python
plt.errorbar(x = bin_c, y = comb_mean);
```

- We can compute the standard deviation of the fuel efficiencies in each bin, and then plot them through the **yerr** parameter.

```python
comb_std = fuel_econ.groupby(displ_bins)['comb'].std()
```

```python
plt.errorbar(x = bin_c, y = comb_mean, yerr= comb_std);
```

- This plot shows that while the cars with smaller engine sizes have generally higher fuel efficiencies, there is also a lot more variability in those efficiencies.

```python
plt.figure(figsize = (15,5))
#Adapted histogram
plt.subplot(1,2,1)
bin_edges = np.arange(0.6, 7.1 + 0.4, 0.4)
displ_bins = pd.cut(fuel_econ['displ'], bins = bin_edges, right = False, include_lowest = True, labels = False)
pts_per_bin = fuel_econ.groupby(displ_bins).size()
comb_wt = fuel_econ['comb'] / pts_per_bin[displ_bins].values
plt.hist(data = fuel_econ, x = 'displ', bins = bin_edges, weights = comb_wt)
plt.title("Adapted Histogram");

#Line plot
plt.subplot(1,2,2)
bin_edges = np.arange(0.6, 7 + 0.2, 0.2)
bin_c = bin_edges[:-1]+0.1
displ_bins = pd.cut(fuel_econ['displ'], bins = bin_edges, include_lowest = True)
comb_mean = fuel_econ.groupby(displ_bins)['comb'].mean()
plt.errorbar(x = bin_c, y = comb_mean)
plt.title("Line Plot");
```
